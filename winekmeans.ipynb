```python
features = df_wine.columns
X_train = df_wine[features].values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(X_train)
```

```python
x_train
```

```python
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters=3)
kmean
```

```python
y_cluster = kmean.fit_predict(x_train)
```

```python
y_cluster
```

```python
df_wine['cluster'] = y_cluster
df_wine
```

```python
kmean.cluster_centers_
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x_train[:,0],x_train[:,1], c=kmean.labels_)
plt.scatter(kmean.cluster_centers_[0,0],kmean.cluster_centers_[0,1],
cmap='rainbow',marker='*',s=150,label='Centroid 0')
plt.scatter(kmean.cluster_centers_[1,0],kmean.cluster_centers_[1,1],
cmap='rainbow',marker='*',s=150,label='Centroid 1')
plt.scatter(kmean.cluster_centers_[2,0],kmean.cluster_centers_[2,1],
cmap='rainbow',marker='*',s=150,label='Centroid 2')
plt.legend()
```

```python
inertias = []
k_range = range(1,10)
for k in k_range:
     km = KMeans(n_clusters=k).fit(x_train)
     inertias.append(km.inertia_)
```

```python
inertias
```

```python
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertias, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')  # Tambahkan argumen untuk xlabel
plt.ylabel('Sum of Squared Errors (Inertia)')
plt.title('Elbow Method')
plt.grid(True)
plt.show()
```

```python
df_wine['cluster'] = df_wine['cluster'].replace([0,1,2],['Tinggi','Sedang','Rendah'])
```
